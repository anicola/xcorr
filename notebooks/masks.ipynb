```python
# from pixell import enmap, utils, powspec, enplot, reproject
import healpy as hp

# to use matplotlib with jupiter
import matplotlib as mpl
mpl.rc('text', usetex=False)
mpl.rc('font', family='serif')
%matplotlib inline
```

```python
# Footprint mask
pathFootMask = "/global/cscratch1/sd/sferraro/SOxpipe/masks/mask_LSST_SO_40pc_nside1024.fits"
footMask = hp.read_map(pathFootMask)

# The mask is in gal coord, so rotate it to eq
hp.mollview(footMask, title="Footprint mask (galactic)")
hp.mollview(footMask, title="Footprint mask (equatorial)", coord=['G', 'C'])
```

```python
pathPsMask = "/global/cscratch1/sd/sferraro/SOxpipe/masks/SO_LSS_ptsources.fits"
psMask = hp.read_map(pathPsMask)
hp.mollview(psMask, title="Point source mask (gal)")
hp.mollview(psMask, title="Point source mask (equa)", coord=['G', 'C'])
```

```python
# Downgrade PS mask to 1024
psMask = hp.ud_grade(psMask, 1024)
# hp.get_nside(psMask)
```

```python
# Combined footprint + point sources mask
psFootMask = psMask * footMask
hp.mollview(psFootMask, title="PS + Foot (gal)")
hp.mollview(psFootMask, title="PS + Foot (equa)", coord=['G', 'C'])

# save it to file
pathPsFootMask = "/global/cscratch1/sd/sferraro/SOxpipe/masks/mask_footLSSTSO40pc_ps_nside1024.fits"
hp.write_map(pathPsFootMask, psFootMask)
```
